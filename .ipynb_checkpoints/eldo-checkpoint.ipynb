```python
import pandas as pd
import numpy as np
import json
```

```python
df=pd.read_json("https://raw.githubusercontent.com/onaio/ona-tech/master/data/water_points.json")
df
```

```python
from collections import Counter
```

```python
#counting the number of water points that are functional
functional= df.water_point_condition.value_counts()['functioning']
print("THE NUMBER OF WATER POINTS THAT ARE FUNCTIONING", functional)

#the new added points
new_points= df.water_point_condition.value_counts()['newly_constructed']
print("THE NUMBER OF NEW WATER POINTS  ",new_points )

#ASSUMING THE NEW POINTS ARE FUNCTIONING
functioning_points=functional+new_points
print("THE NUMBER OF WATER POINTS THAT ARE FUNCTIONING + NEW POINTS",functioning_points )
```

```python
#the number of water points per community
community= df.respondent.value_counts()['community']
print("THE NUMBER OF WATER POINTS IN community",community)

local_authority_community= df.respondent.value_counts()['local_authority community']
print("THE NUMBER OF WATER POINTS IN local_authority_community",local_authority_community)
```

```python
filter_criteria =(df['respondent'] == 'community' )& (df['water_point_condition'] == 'broken')
df1m = df[filter_criteria]

print(df1m.shape)
#counting the community water point is broken
community= df1m.water_point_condition.value_counts()['broken']

print('total number of broken points in %:', community/100)


df1m.head(10)
```

```python
filter_criteria =(df['respondent'] == 'local_authority community' )& (df['water_point_condition'] == 'broken')
df3m = df[filter_criteria]

print(df3m.shape)
#counting the community water point is broken
local_authority_community = df3m.water_point_condition.value_counts()['broken']

print('total number of broken points in %:',local_authority_community /100)


df3m.head(10)
```

```python
filter_criteria =(df['respondent'] == 'local_authority' )& (df['water_point_condition'] == 'broken')
df2m = df[filter_criteria]

print(df2m.shape)
#counting the community water point is broken
local_authority = df2m.water_point_condition.value_counts()['broken']

print('total number of broken points in %:', local_authority/100)


df2m.head(10)
```
